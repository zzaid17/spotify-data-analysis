Import libraries, read and clean data, and set up data for plotting:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data.csv")
df.drop_duplicates(inplace=True)

df['Popularity'] = pd.to_numeric(df['Popularity'])
df['Duration (s)'] = (pd.to_numeric(df['Duration (ms)'])) / 1000
```

Define and plot popularity vs duration:

```python
def plot_duration_popularity():
  df.plot(x='Duration (s)', y='Popularity', kind='scatter')
  plt.xlabel('Duration (s)')
  plt.ylabel('Popularity')
  plt.title('Spotify song popularity vs duration')

plot_duration_popularity()
plt.tight_layout()
plt.show()
```

Define and plot artist popularity

```python
def plot_artist_popularity():
  artist_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
  top_20 = artist_popularity.head(30)
  top_20.plot(kind='bar')
  plt.xlabel('Artist')
  plt.ylabel('Popularity')
  plt.title('Artist Popularity')
  plt.tick_params(axis='x', rotation=45)

plot_artist_popularity()
plt.tight_layout()
plt.show()
```
